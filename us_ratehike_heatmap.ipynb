```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
#元データ取得
data = pd.read_csv('https://raw.githubusercontent.com/miotomita/us_ratehike/main/heatmap_data.csv',index_col=0).applymap(lambda x:pd.to_numeric(x)).astype(float)
```

```python
data.index = pd.to_datetime(data.index).strftime('%Y/%m/%d')
```

```python
data = data.rename(columns={
    '住宅価格の上昇率（前年同月比）':'住宅価格',
    '雇用コスト指数の伸び率':'雇用コスト',
    '消費者物価の上昇率':'消費者物価',
    '消費者物価の上昇率（食品・エネルギー除く）':'消費者物価（食品・エネルギー除く）',
    '卸売り物価（PPI）の上昇率':'卸売物価',
    '個人消費の伸び':'個人消費',
    '新規住宅着工件数':'新規住宅着工'})
```

```python
#表示項目・順番入れ替え
align = ['経済成長率','失業率','雇用コスト','インフレ期待','消費者物価','消費者物価（食品・エネルギー除く）','卸売物価','個人消費','小売り在庫／販売の比率','住宅価格','新規住宅着工']
```

```python
data = data.loc[:,align]
```

データ調整：外れ値を削る（ここでは-3, 3で指定）
色の見栄え次第で数字は動かしていただいても構いません・・・が、だいたい＋ー３がコロナ前後の異常値をはじける水準でした

```python
#上・下を＋３、ー３で固定
source = np.clip(data,-3,3)
```

①プリフィクスのカラーレンジ、その１
<br>この中からどれでも指定可能<br>
https://vega.github.io/vega/docs/schemes/

```python
colorscheme = 'blueorange'
```

```python
#色味、プリフィクスのカラースキーム
heatmap1 = alt.Chart(
    source.reset_index()).transform_fold(
    source.columns.tolist(),as_=['Indices','Z-Score']).mark_rect(stroke='white', strokeWidth=1).encode(
    x='date:O',
    y=alt.Y('Indices:O',sort=align),
    color=alt.Color('Z-Score:Q', scale = alt.Scale(scheme=colorscheme))
                   ).properties(height=360,width=360)
heatmap1
```

②プリフィクスのカラーレンジ、その２（減速・加速の色の都合上、反転が必要なスキーム）
<br>この中からどれでも指定可能<br>
https://vega.github.io/vega/docs/schemes/

```python
colorscheme = 'redyellowblue'
```

```python
#色味、プリフィクスのカラースキームを反転させる場合
#凡例は数字が逆に出るので調整が必要
source2 = source * -1


heatmap2 = alt.Chart(
    source2.reset_index()).transform_fold(
    source2.columns.tolist(),as_=['Indices','Z-Score']).mark_rect(stroke='white', strokeWidth=1).encode(
    x='date:O',
    y=alt.Y('Indices:O',sort=align),
    color=alt.Color('Z-Score:Q', scale = alt.Scale(scheme=colorscheme))
                   ).properties(height=360,width=360)
heatmap2
```

```python
#７色
colorrange = ['#0C008C','#0009FF','#99CCFF','#FFFFFF','#FFEA80','#FFAA00','#804000']
```

```python
heatmap3 = alt.Chart(
    source.reset_index()).transform_fold(
    source.columns.tolist(),as_=['Indices','Z-Score']).mark_rect(stroke='white', strokeWidth=1).encode(
    x='date:O',
    y=alt.Y('Indices:O',sort=align),
    color=alt.Color('Z-Score:Q', scale = alt.Scale(range=colorrange, domainMid=0))
                   ).properties(height=360,width=360)
heatmap3
```
